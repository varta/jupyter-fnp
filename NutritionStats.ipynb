```python
import re
from lxml import html
import pandas as pd

def parse_data(r):
    tree = html.fromstring(r)
    
    
    title = tree.xpath('//title')
    if title == None or len(title) < 1:
        raise ValueError 
    title = title[0].text_content()
    title = title[:title.find(' Nutrition Facts')].strip('\n ')
    
    tables = tree.xpath('//table')
    if tables == None or len(tables) < 1:
        raise ValueError
    tables = tables[-1].text_content()
    
    nutrients = tree.xpath('//div[@class="clearer"]//span')
    if nutrients == None or len(nutrients) < 1:
        raise ValueError
    nutrients = [i.text_content() if i.values()[0] == 'indented' else i.values()[0] 
                 for i in nutrients]
    idx = 0
    facts = []
    while idx < len(nutrients):
        if nutrients[idx].startswith('NUTRIENT_'):
            idx += 3
            continue
        facts += (nutrients[idx:idx+4], )
        idx += 4

    values = None
    script_begin = 'foodNutrients = {'
    script_end = '};'
    for i in tree.xpath('.//script[@type="text/javascript"]'):
        if i.text_content().find(script_begin) > 0:
            values = i.text_content()
            values = values[values.find(script_begin)+len(script_begin):values.find(script_end)].split('",')
            stripper = lambda xx: [x.strip(' "') for x in xx.split(':')]
            values = [tuple(stripper(v)) for v in values]
            #floaterr = lambda xx: None if set(xx).intersection(set('0123456789.-')) == {} else float(xx)
            def floaterr(xx):
                try:
                    return float(xx)
                except:
                    return None
            values = {iv[0]: iv[1] if iv[0].startswith('UNIT_NUTRIENT_')
                      else (floaterr(iv[1]) if iv[0].startswith('NUTRIENT_') or iv[0].startswith('DV_NUTRIENT_') else None)
                      for iv in values}
            break
            
    nutrition = [('Nutrient', title)]
    dctlookup = lambda xx, dct: values[xx] if xx is not None and xx in dct else None
        
    div = lambda a, b: a*100.0/b if a is not None and b is not None else None
    for ifacts in facts:
        nutrition += [(ifacts[0], dctlookup(ifacts[1], values)),
                      (ifacts[0]+'_units', dctlookup(ifacts[2], values)),
                      (ifacts[0]+'%', div(dctlookup(ifacts[1], values), dctlookup(ifacts[3], values)))]

    with pd.option_context('display.max_rows', None, 'display.max_columns', None):
        df = pd.DataFrame(nutrition).T
        df.columns = df.loc[0]
        df = df.drop([0])
        
    return title, tables, nutrients, facts, values, nutrition, df, r, tree

def get_nutrition_food(url):
    with urlopen(url) as response:
        response_content = response.read().decode()
    return parse_data(response_content)

def get_table(ns):
    with pd.option_context('display.max_rows', None, 'display.max_columns', None,
                           'display.max_colwidth', None): 
        #pd.set_option('display.max_colwidth', None)
        cols = ['Potassium%', 'Phosphorus%', 'Protein%', 'Calcium%', 'Sodium%', 'Vitamin A%', 'Vitamin C%']
        if 'Nutrient' in ns.columns:
            cols = ['Nutrient'] + cols
        display(ns[cols])

    
def get_table_file_(file='nutrition_stats.tsv'):
    return pd.read_csv(file, sep='\t')
    
def get_table_url_(url, file='nutrition_stats.tsv'):
    title, tables, nutrients, facts, values, nutrition, df, r, tree = get_nutrition_food(url)
    nutritionstats = get_table_file_(file)
    nutritionstats = nutritionstats.append(df)
    nutritionstats.to_csv(path_or_buf=file, sep='\t')    
    return nutritionstats

def fixindex(ns):
    if 'Nutrient' in ns.columns:
        ns.index = ns.Nutrient
        ns = ns.drop('Nutrient', axis=1)
    return ns

def get_table_file(file='nutrition_stats.tsv'):
    return fixindex(get_table_file_(file))

def get_table_url(url, file='nutrition_stats.tsv'):
    ns = fixindex(get_table_url_(url, file))
    get_table(ns)
    return ns
```
